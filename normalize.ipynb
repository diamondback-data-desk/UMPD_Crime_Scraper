## Import statements

```python
from geopy.geocoders import Nominatim
import pandas as pd
import folium
import time
```

```python
crime_export_df = pd.read_csv("UMPD_crime_export_1_2017_to_11_2024.csv")
crime_export_df = crime_export_df[crime_export_df["case_number"].str[:4] == "2024"]

for index,row in crime_export_df.iterrows():
    location_desc = f"{row["location_desc"]}, COLLEGE PARK, MARYLAND"
    row["location_desc"] = location_desc
    crime_export_df.loc[index] = row
    
crime_export_df
```

```python
crime_analysis_df = crime_export_df[["case_number","date_reported","time_reported","type","disposition"]]
crime_analysis_df
```

```python
crime_export_df["norm_loc_desc"] = None
crime_export_df["latitute"] = None
crime_export_df["longitude"] = None

for index,row in crime_export_df.iterrows():
    geolocator = Nominatim(user_agent="UMPD_Location_Normalizer")
    location = geolocator.geocode(row["location_desc"])
    print(location)
    if location:
        row["norm_loc_desc"] = location
        row["latitude"] = location.latitude
        row["latitude"] = location.longitude
        crime_export_df.loc[index] = row
        time.sleep(1)
    
crime_export_df
```
